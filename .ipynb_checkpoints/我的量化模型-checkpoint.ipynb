```python
import tushare as ts
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import talib
%matplotlib inline
```

```python
GP = 'sz50'
```

```python
data = ts.get_hist_data(str(GP))
```

```python
print data
```

```python
data1 = data.values

open = data1[:,0]
high = data1[:,1]
close = data1[:,2]
low = data1[:,3]
volume = data1[:,4]
```

```python
#复制data表结构，重建一个新的指标表
ml_datas = data.drop(data.columns,axis=1)


macd,macdsignal,macdhist = talib.MACD(close, fastperiod=12, slowperiod=26, signalperiod=9)
obv = talib.OBV(close,volume)
wcl = talib.WCLPRICE(high,low,close)

time = [5,10,20,30,60]
for i in time:
    locals()['sma'+str(i)] = talib.SMA(close, timeperiod=i)
    locals()['rsi'+str(i)] = talib.RSI(close, timeperiod=i)
    locals()['roc'+str(i)] = talib.ROC(close, timeperiod=i)
    locals()['mom'+str(i)] = talib.MOM(close, timeperiod=i)
    locals()['atr'+str(i)] = talib.ATR(high,low,close, timeperiod=i)
    locals()['cci'+str(i)] = talib.CCI(high,low,close, timeperiod=i)
    locals()['bbandupper'+str(i)],locals()['bbandmiddle'+str(i)],locals()['bbandlower'+str(i)] = talib.BBANDS(close, timeperiod=i)

#将指标加入原数据库
for i in time:
    ml_datas['sma'+str(i)] = pd.Series(locals()['sma'+str(i)],index = data.index)
    ml_datas['rsi'+str(i)] = pd.Series(locals()['rsi'+str(i)],index = data.index)
    ml_datas['roc'+str(i)] = pd.Series(locals()['roc'+str(i)],index = data.index)
    ml_datas['mom'+str(i)] = pd.Series(locals()['mom'+str(i)],index = data.index)
    ml_datas['atr'+str(i)] = pd.Series(locals()['atr'+str(i)],index = data.index)
    ml_datas['cci'+str(i)] = pd.Series(locals()['cci'+str(i)],index = data.index)
    ml_datas['bbandupper'+str(i)] = pd.Series(locals()['bbandupper'+str(i)],index = data.index)
    ml_datas['bbandmiddle'+str(i)] = pd.Series(locals()['bbandmiddle'+str(i)],index = data.index)
    ml_datas['bbandlower'+str(i)] = pd.Series(locals()['bbandlower'+str(i)],index = data.index)
ml_datas['macd'] = pd.Series(macd,index = data.index)
ml_datas['macdsignal'] = pd.Series(macdsignal,index = data.index)
ml_datas['macdhist'] = pd.Series(macdhist,index = data.index)
ml_datas['obv'] = pd.Series(obv,index = data.index)
ml_datas['wcl'] = pd.Series(wcl,index = data.index)


#将原始数据集的数据移动一天，使每天收盘价数据的特征训练的时候用前一天的信息
ml_datas = ml_datas.shift(1)

ml_datas['reg_target'] = data['close']
ml_datas['clf_target'] = (data['close'] / data['close'].shift(1))-1>0
```

```python
#删除空行
ml_datas = ml_datas.dropna()
```

```python
X_ori = ml_datas.drop(['reg_target','clf_target','obv'],axis=1)
X_ori.describe()
```

```python
y = ml_datas['clf_target']
y.describe()
```

```python
#特征标准化
from sklearn import preprocessing

scaler = preprocessing.StandardScaler().fit(X_ori)

X = scaler.transform(X_ori)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

forest = ExtraTreesClassifier(n_estimators = 250,random_state = 0)
forest.fit(X,y)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],axis=0)
indices = np.argsort(importances)[::-1]
print('Feature ranking')

for f in range(len(indices)):
    print('%d, feature %s (%f)' % (f+1,X_ori.columns[indices[f]],importances[indices[f]]))

indices = indices[:20]
```

```python
plt.figure(figsize=(16,9))
plt.title('Feature Importance')
plt.bar(range(len(indices)),importances[indices],color='r',yerr=std[indices],align='center')
plt.xticks(range(len(indices)),indices)
plt.xlim([-1,len(indices)])
plt.show()
```
