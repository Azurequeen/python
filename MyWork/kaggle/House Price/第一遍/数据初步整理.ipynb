- **处理缺失值**<br>
鉴于该数据集质量较高，缺失值不做drop，分为数值型和字段型分别补齐<br>
- 字段型：用一个字段替换
- 数值型：替换成一个值

```python
#查看有缺失的列
df1 = data_train1.isnull()
df2 = df1.apply(lambda x: x.sum(), axis=0)
df2[df2 > 0]
```

```python
#字符型处理较为简单，用一个新特征代表缺失
#门口铺设，Alley，NaN 代表 No alley access，替换为 NoAlley
#砌体类型，MasVnrType，NaN代表无砌体
#地下室高度、状况、光照、评级、第二地下室评级，BsmtQual、BsmtCond、BsmtExposure、BsmtFinType1、BsmtFinType2，NaN 代表 No Basement，替换为 NoBasement
#游泳池质量，PoolQC，NA 代表 No Pool
#栅栏质量，Fence，NA 代表 No Fence
#杂项功能，MiscFeature，NA 代表 None
#车库类型 、建成时间 、装饰、质量、条件，GarageType、GarageFinish、GarageCars、GarageArea、GarageQual NA代表No Garage
#壁炉质量，FireplaceQu，NA 代表 No Fireplace


data_train1.loc[ (data_train1['门口铺设'].isnull()), '门口铺设' ] = 'NoAlley'
print '门口铺设修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['Alley'].isnull().sum(),data_train1['门口铺设'].isnull().sum())

data_train1.loc[ (data_train1['砌体类型'].isnull()), '砌体类型' ] = 'NoMas'
print '砌体类型修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['MasVnrType'].isnull().sum(),data_train1['砌体类型'].isnull().sum())

data_train1.loc[ (data_train1['地下室高度'].isnull()), '地下室高度' ] = 'NoBasement'
print '地下室高度修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['BsmtQual'].isnull().sum(),data_train1['地下室高度'].isnull().sum())

data_train1.loc[ (data_train1['地下室状况'].isnull()), '地下室状况' ] = 'NoBasement'
print '地下室状况修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['BsmtCond'].isnull().sum(),data_train1['地下室状况'].isnull().sum())

data_train1.loc[ (data_train1['地下室光照'].isnull()), '地下室光照' ] = 'NoBasement'
print '地下室光照修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['BsmtExposure'].isnull().sum(),data_train1['地下室光照'].isnull().sum())

data_train1.loc[ (data_train1['地下室评级'].isnull()), '地下室评级' ] = 'NoBasement'
print '地下室评级修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['BsmtFinType1'].isnull().sum(),data_train1['地下室评级'].isnull().sum())

data_train1.loc[ (data_train1['第二地下室评级'].isnull()), '第二地下室评级' ] = 'NoBasement'
print '第二地下室评级修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['BsmtFinType2'].isnull().sum(),data_train1['第二地下室评级'].isnull().sum())

data_train1.loc[ (data_train1['游泳池质量'].isnull()), '游泳池质量' ] = 'NoPool'
print '游泳池质量修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['PoolQC'].isnull().sum(),data_train1['游泳池质量'].isnull().sum())

data_train1.loc[ (data_train1['栅栏质量'].isnull()), '栅栏质量' ] = 'NoFence'
print '栅栏质量修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['Fence'].isnull().sum(),data_train1['栅栏质量'].isnull().sum())

data_train1.loc[ (data_train1['杂项功能'].isnull()), '杂项功能' ] = 'NoMisc'
print '杂项功能修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['MiscFeature'].isnull().sum(),data_train1['杂项功能'].isnull().sum())

data_train1.loc[ (data_train1['车库类型'].isnull()), '车库类型' ] = 'NoGarage'
print '车库类型修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['GarageType'].isnull().sum(),data_train1['车库类型'].isnull().sum())

data_train1.loc[ (data_train1['车库建成时间'].isnull()), '车库建成时间' ] = 'NoGarage'
print '车库建成时间修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['GarageFinish'].isnull().sum(),data_train1['车库建成时间'].isnull().sum())

data_train1.loc[ (data_train1['车库装饰'].isnull()), '车库装饰' ] = 'NoGarage'
print '车库装饰修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['GarageCars'].isnull().sum(),data_train1['车库装饰'].isnull().sum())

data_train1.loc[ (data_train1['车库质量'].isnull()), '车库质量' ] = 'NoGarage'
print '车库质量修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['GarageArea'].isnull().sum(),data_train1['车库质量'].isnull().sum())

data_train1.loc[ (data_train1['车库条件'].isnull()), '车库条件' ] = 'NoGarage'
print '车库条件修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['GarageQual'].isnull().sum(),data_train1['车库条件'].isnull().sum())

data_train1.loc[ (data_train1['壁炉质量'].isnull()), '壁炉质量' ] = 'NoFire'
print '壁炉质量修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['FireplaceQu'].isnull().sum(),data_train1['壁炉质量'].isnull().sum())

```

```python
#数值型处理需要对照字段一一处理
#临街距离，LotFrontage，说明文件中未描述，但是根据分布，猜测小于20的都是NaN，所以全部补0
data_train1.loc[ (data_train1['临街距离'].isnull()), '临街距离' ] = 0
print '修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['LotFrontage'].isnull().sum(),data_train1['临街距离'].isnull().sum())

#胶合板面积，MasVnrArea，明文件中未描述，猜测全部补0
data_train1.loc[ (data_train1['胶合板面积'].isnull()), '胶合板面积' ] = 0
print '修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['MasVnrArea'].isnull().sum(),data_train1['胶合板面积'].isnull().sum())

```

```python
#电气系统，Electrical，只有1个缺失，丢弃
data_train1 = data_train1.dropna(subset=['电气系统']) 
print '电气系统修改前NaN值为%s个,修改后NaN值为%s个' % (data_train['Electrical'].isnull().sum(),data_train1['电气系统'].isnull().sum())
```

```python
#查看处理完缺失后的情况
df1 = data_train1.isnull()
df2 = df1.apply(lambda x: x.sum(), axis=0)
print '处理完毕，缺失值为%s' % df2[df2 > 0].count()
```
