## 生成文件名npy

```python
orig = os.listdir(pic_path)
full = []
for i in orig:
    if '_clean' in i:
        full.append(i[:-10])
```

```python
random.shuffle(full)
lenth = len(full)
kaggleluna_full = full[0:int(lenth*0.8)]
valsplit = full[:-int(lenth*0.2)]
np.save(pic_path+'kaggleluna_full.npy',kaggleluna_full)
np.save(pic_path+'valsplit.npy',valsplit)
```
